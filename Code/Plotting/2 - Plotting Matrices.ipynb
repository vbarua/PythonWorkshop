````python
import matplotlib.pyplot as plt
import numpy as np

# For presentation purposes only.
%matplotlib inline
````

# Plotting Matrices
At some point you may need to plot a matrix. This works a bit differently from regular plots.

````python
def create_matrix(size):
    mat = np.zeros((size, size))
    for i in range(size):
        for j in range (size):
            mat[i, j] = i * j
    return mat
````

````python
mat = create_matrix(20)
plt.imshow(mat)
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matrix Plot - With Interpolation")
````

The plot above looks very smooth even though it shouldn't. This is because matplotlib will by default interpolate the values of the matrix. Let's plot the same matrix with interpolation turned off.

````python
mat = create_matrix(20)
plt.imshow(mat, interpolation='none') # Interpolation turned off.
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matrix Plot - Without Interpolation")
````

With interpolation turned off the true structure of the data is revealed. Upping the density of the matrix can help produce a better plot even with interpolation removed.

````python
mat = create_matrix(100)
plt.imshow(mat, interpolation='none')
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Denser Matrix Plot - Without Interpolation")
````

The color scale used to represent the data can also be modified using the ```cmap``` keyword argument.

````python
import matplotlib.cm as cm

plt.imshow(mat, cmap=cm.Reds, interpolation='none')
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Switching Color Scale ")
````

There are lots of different colormaps to choose from.

````python
plt.imshow(mat, cmap=cm.winter_r, interpolation='none')
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Switching Color Scale ")
````
